Normally, when dealing with plain Python objects, we'd use len() to get the size of a list. For NumPy arrays that gives us only the size of the first dimension, so it's not very useful. Instead, use shape to get the dimensions of the array as a tuple

```python
len(im)
```

```python
im.shape
```

There's also a property called size that contains the number of elements:

```python
im.size
```

```python
im1 = imread('C:/Users/West/OneDrive/Desktop/sushi1.png')
im2 = imread('C:/Users/West/OneDrive/Desktop/sushi2.png ')
```

It would be helpful to display those two images next to each other; we can do that with subplots:

```python
fig, ax = plt.subplots(1, 2)
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].imshow(im1)
ax[1].imshow(im2)
```

We can add titles to the subplots but also to the figure itself:

```python
fig, ax = plt.subplots(1, 2)
ax[0].imshow(im1)
ax[1].imshow(im2)

ax[0].set_title("original")
ax[1].set_title("modified")
fig.suptitle("Find the difference")
```
