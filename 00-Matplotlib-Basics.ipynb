## Visualizing Plots

Review video to understand this section!

```python
import matplotlib.pyplot as plt

# JUPYTER NOTEBOOK ONLY
# %matplotlib inline
```

The first argument to the plot() function call is the list of values for X-axis, the second argument is the list of values for Y-axis, and the third argument is the visualization style. It will automatically create the following output

```python
x = [0,1,2]
y = [100,200,300]
```

```python
plt.plot(x,y)
```

```python
x = np.arange(5)
y=x
```

```python
plt.plot(x, y, 'o--')
plt.plot(x, -y, 'o-')
plt.show()
```

We can use title() function to add title to the visualization. Run the following code and see the output

```python
plt.plot(x, y, 'o--')
plt.plot(x, -y, 'o-')
plt.title('y=x and y=-x')
plt.show()
```

```python
plt.plot(x,y)
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

```python
plt.plot(x,y)

# Axis and ticks
plt.xlim(0,2)
plt.ylim(0,10)


# Labeling
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

```python
plt.plot(x,y,color='red')

# Axis and ticks
plt.xlim(0,2)
plt.ylim(0,10)


# Labeling
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

```python
plt.plot(x,y,color='red',marker='o',markersize=20,linestyle='--')

# Axis and ticks
plt.xlim(0,2)
plt.ylim(0,10)


# Labeling
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

```python
plt.plot(x, y, 'o--')
plt.axis ('off')
plt.show()
```

# Scatter Plot

Adding multiple lines and a legend

```python
x=np.linspace(0,10,50)
y1=np.power(x,2)
y2=np.power(x,3)
plt.plot(x,y1,'b.',label='$x^2$')
plt.plot(x,y2,'go', label='$x^3$')
plt.xlim((1,5))
plt.ylim((0,30))
plt.xlabel('my x label')
plt.ylabel('my y label')
plt.title("plot title")
plt.legend()
plt.savefig('cutve_plot.pdf')
```

# Histogram

```python
data=np.random.randn(1000)
f,(ax1,ax2)=plt.subplots(1,2,figsize=(6,3))
ax1.hist(data, bins=30, normed=True, color='b')
ax2.hist(data, bins=30, normed=True, color='r', cumulative=True)
plt.savefig('histogram.pdf')
```
